## 1NN and 3NN Classification and Evaluation

```python
data_norm["Labels"] = np.where(data_norm["Labels"].str.contains("M"), 1, 0)
```

```python
Y = data_norm['Labels'].to_numpy()
X = data_norm.iloc[:,0:30].to_numpy()
```

```python
data_norm.head()
```

### Using leave one out cross validation for calculating the accuracy of 1NN and 3NN

```python
loo = LeaveOneOut()
```

```python
KNN_1 = KNeighborsClassifier(n_neighbors=1)
```

```python
KNN_3 = KNeighborsClassifier(n_neighbors=3)
```

```python
accuracy_1nn = cross_val_score(KNN_1, X, Y, scoring='accuracy', cv = loo)
print('Accuracy of the 1NN classifier: %.3f (%.3f)' % (np.mean(accuracy_1nn), np.std(accuracy_1nn)))
```

```python
accuracy_3nn = cross_val_score(KNN_3, X, Y, scoring='accuracy', cv = loo)
print('Accuracy of the 3NN classifier: %.3f (%.3f)' % (np.mean(accuracy_3nn), np.std(accuracy_3nn)))
```

As expected 3NN performs better than the 1NN algorithm
